## Tensorflow Denemeleri

Geçtiğimiz haftanın büyük bir bölümünde, Derin Öğrenme modellerinizde tensörler kullandınız. Bugünkü zorluklar boyunca, tensörleri kapsamlı bir şekilde kullanacak ve bazı alışılmadık model mimarileriyle çalışacaksınız. 


Bu ısınma çalışmasında, bazı yeni yöntemlerle tanışacaksınız. 

Buradaki tüm yöntemler, günün ilerleyen saatlerinde karşılaşacağınız zorlukları 100 kat daha kolay hale getirecek, bu nedenle daha sonra takılırsanız, doğru cevapları nasıl bulduğunuzu görmek için buraya geri dönün. 

```python
import tensorflow as tf
```

```python

import numpy as np
```

A) `tf.ones` kullanarak (3,3) şeklinde birler ile doldurulmuş bir tensör `a` oluşturun.

```python
# (3, 3) boyutunda, içi 1 dolu tensör
a = tf.ones((3, 3))
print(a)
```

B) tf.expand_dims kullanarak `a`'yı `b` adlı (1,3,3) şekilli bir nesne haline getirin.

```python
# a tensörünün başına (0. indeks) yeni bir boyut ekliyoruz
b = tf.expand_dims(a, axis=0)
print(b.shape)
```

C) Boyutu (9,1) olan ve sıfırlarla dolu bir tensör `c` oluşturun.

```python
# (9, 1) boyutunda, içi 0 dolu tensör
c = tf.zeros((9, 1))
print(c)
```

D) `c`'yi (1, 3, 3) şeklinde yeniden şekillendirin. Adını `d` olarak belirleyin.

```python
# c tensörünü al, (1, 3, 3) şekline sok
d = tf.reshape(c, (1, 3, 3))
print("Eski şekil:", c.shape)
print("Yeni şekil:", d.shape)
```

E) `tf.matmul` kullanarak `b` ve `d` matrislerini birbiriyle çarpın ve şimdi tensörü `e`'ye atayın. 

```python
# b ve d matrislerini çarp
e = tf.matmul(b, d)
print(e)
```

F) Aşağıda bir `numpy` dizisi bulunmaktadır. Bunu `array_tensor` adlı bir tensöre dönüştürün. Ardından `tf.cast` kullanarak bunun `int` değil `float` türü olduğundan emin olun. Sonucu `f`'ye atayın.

```python
import numpy as np
```

```python
array = np.array([[
    [1,2,3],
    [1,2,3],
    [4,5,6],
    [7,8,9]
]])
```

```python
# 1. Önce NumPy dizisini Tensor'a çevirelim
array_tensor = tf.convert_to_tensor(array)

# 2. Veri tipini integer'dan float'a (ondalıklı) çevirelim
f = tf.cast(array_tensor, dtype=tf.float32)

print("Veri Tipi:", f.dtype)
print("Şekil:", f.shape)
```

G) `f`'nin __last column__ değerleri seçin ve bunu `g`'ye kaydedin. Açıkça belirtmek gerekirse, `g`'nin şekli (4,) olmalı ve `3,3,6,9` değerlerini içermelidir. 

```python
# Tensörün içinden [ilk paket, tüm satırlar, son sütun] verisini çek
g = f[0, :, -1]

print("Sonuç:", g)
print("Beklenen Şekil (4,) mi?:", g.shape)
```

H) g boyutlarını (4,1) şekline gelecek şekilde genişletin. Bunu `h`'ye atayın.

```python
# g vektörünü (4, 1) boyutuna getiriyoruz
h = tf.expand_dims(g, axis=-1)

print("Eski Şekil:", g.shape)
print("Yeni Şekil:", h.shape)
```

I) İsteğe bağlı `transpose_b` argümanıyla `tf.matmul` kullanarak e ile f'yi çarpın (konumsal argümanlarınızda önce `e`, sonra `f` gelmelidir). Bunu `i`'ye atayın.

```python
# e ile f'yi çarpacağız ama f'nin transpozunu alarak (transpose_b=True)
i = tf.matmul(e, f, transpose_b=True)

print("Sonuç Şekli:", i.shape)
print(i)
```

J) Birlerle dolu bir `(10,10)` tensör oluşturun. Ardından `tf.linalg.band_part()` kullanarak matrisin üst üçgenini maskeleme yapın (yani üst üçgenin tamamı 0 olmalıdır). `j`'ye atayın.

```python
# 1. Önce 10x10'luk, içi 1 dolu bir matris yapalım
ones_matrix = tf.ones((10, 10))

# 2. Alt tarafı tut (-1), üst tarafı sil (0)
# Bu işlem üst üçgeni 0 yapar.
j = tf.linalg.band_part(ones_matrix, num_lower=-1, num_upper=0)

print(j)
```

K) Aşağıdaki `number_list`'i alın ve bir tensör haline getirin. Bunu `k`'ye atayın.

```python
number_list = [1,2,3,4,8]
```

```python
# Listeyi tensöre çevir
k = tf.convert_to_tensor(number_list)

print(k)
```

L) `tf.expand_dims` kullanarak `k`'yi (1,5) şekilli bir tensör haline getirin. `l`'ye atayın.

```python
# k tensörüne "satır" boyutu ekliyoruz (en başa)
l = tf.expand_dims(k, axis=0)

print("k'nın şekli:", k.shape)
print("l'nin şekli:", l.shape)
```

M) `tf.tile` kullanarak `l` (okunuşu “L”) 50 kez tekrarlayın, sonuçta `(50, 5)` şeklinde olsun. `m`'ye atayın.

```python
# [50 kere aşağı, 1 kere yana] kopyala
m = tf.tile(l, multiples=[50, 1])

print("m'nin şekli:", m.shape)
print(m)
```

N) Basit bir Boolean ifadesi kullanarak, `m`'nin 3 olduğu yerler için `True` değerleri ve 3 olmadığı yerler için `False` değerleri içeren bir tensör oluşturun. `n`'ye atayın.

```python
# m tensöründe 3'e eşit olan yerler True olsun
n = (m == 3)

print(n)
# Çıktıda True ve False'lar göreceksin
```

O) Değişkeniniz `m` içindeki her değeri 3'e bölün. `o`'ya atayın.

```python
# Tüm elemanları 3'e böl
o = m / 3

print(o)
```

P) `tf.concat()` kullanarak, aşağıdaki (5,5) ve (5,5) şekilli iki tensörün listesini alın ve bunları (10,5) şekilli son bir tensöre dönüştürün. `p`'ye atayın. Birleştirme eksenini dikkatlice düşünün.

```python
tensor_list = [tf.ones((5,5)), tf.ones((5,5))]
```

```python
# tensor_list zaten tanımlıydı, onu kullanıyoruz
# axis=0 diyerek alt alta yapıştırıyoruz
p = tf.concat(tensor_list, axis=0)

print("Sonuç Şekli:", p.shape)
```
